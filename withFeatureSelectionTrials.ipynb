# Build the ANNOY index Using all-MiniLM-L6-v2

```python
#Using ANNOY: Approximate Nearest Neighbours Oh yeah! by Spotify for Type1s
from annoy import AnnoyIndex

import time
start = time.time()

f = 384

t_ctr_txt_1 = AnnoyIndex(f, 'angular')  #Length of item vector that will be indexed
#for index, row in dummy_df.head(1000):
   # t.add_item(index, row)
#adf = final_df.iloc[:500000, :]
for index, row in ctr_embeds_df_1.iterrows():
    #print(f"index: {index}, row: {row.nif}")
    t_ctr_txt_1.add_item(index, row)

t_ctr_txt_1.build(10) # 10 trees
t_ctr_txt_1.save('test_ctr_txt_1.ann')

end = time.time()

# Total time to compute
print('\nThis function took {:.2f} s to compute.'.format(end - start))   
```

```python
import nltk
import re
from nltk.corpus import stopwords
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
def lemmatize_words(text):
    text = re.sub('[^a-zA-Z]', ' ', text)
    #print(text)
    text = text.lower()
    words = text.split()
    words = [lemmatizer.lemmatize(word,pos='v') for word in words if not word in set(stopwords.words('portuguese'))]
    return ' '.join(words)
samp_ctr['category1s_text_lemm'] = samp_ctr['category1s_text'].apply(lemmatize_words)
```
